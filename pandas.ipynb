```python
# Wrong Value Change Method

df = pd.read_csv('emp.csv')

for i in df.index:
    if df.loc[i, "Salary"] < 50000:
        df.loc[i, "Salary"] = 40000

print(df.to_string())
```

```python
# Wrong Value Drop Method

for i in df.index:
    if df.loc[i, "Salary"] < 50000:
        df.drop(i, inplace = True)

print(df.to_string())
```

# Pandas - Removing Duplicates

```python
import pandas as pd
```

```python
# Removing Duplicates

df=pd.read_csv('emp1.csv',index_col=[0])

print(df.to_string())
print(df.duplicated())
```

```python
df = pd.read_csv('emp1.csv',index_col=[0])

df.drop_duplicates(inplace = True)

print(df.to_string())
```

```python
# Removing Duplicates

df = pd.read_csv('data (1).csv')

print(df.duplicated())
```

```python
df = pd.read_csv('data (1).csv')

df.drop_duplicates(inplace = True)

print(df.to_string())
```
